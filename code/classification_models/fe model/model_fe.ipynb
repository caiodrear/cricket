```python
match_data = pd.read_csv('../../../data/fe_data.csv')
match_data
```

```python
metric_list=['rel_bat_avg','rel_bat_sr']+['bowl_avg','bowl_sr','rel_economy']
metric_list+=[metric+'_form' for metric in metric_list]
metric_list

def preprocess_data(match_data, metrics):
    
    X, y = match_data[metrics], match_data['set_team_win']
    
    undersample = RandomUnderSampler(sampling_strategy = 'majority')
    X, y = undersample.fit_resample(X, y)

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, shuffle = True)

    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = preprocess_data(match_data, metric_list)
```
